```python
import seaborn as sns
```

## Count Plot

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.to_csv('titanic.csv')
```

```python
titanic
```

```python
sns.countplot(data=titanic, x='class')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.countplot(data=titanic, x='class', palette='Oranges')
```

```python
sns.countplot(data=titanic, x='class', hue='who')
```

```python
sns.countplot(data=titanic, x='class', hue='deck')
```

```python
sns.countplot(data=titanic, y='class', hue='deck')
```

```python
ax = sns.countplot(data=titanic, x='class', hue='who')
for element in ax.containers:
    ax.bar_label(element)
```

## heatmap

```python
ax = sns.countplot(data=titanic, x='who', hue='alive')
for element in ax.containers:
    ax.bar_label(element)
```

```python
sns.heatmap(data=titanic[['fare']])
```

```python
data = [[99,15,85],
        [78,5,16],
        [56,2,5]]
```

```python
sns.heatmap(data=data, annot=True)
```

```python
glue = sns.load_dataset("glue").pivot(index="Model", columns="Task", values="Score")
sns.heatmap(glue)
```

```python
sns.heatmap(data=glue, annot=True)
```
